```python
data_cleaned = data_2.copy()
```

```python
###Creating dummies for categorical variables
```

```python
data_with_dummies = pd.get_dummies(data_cleaned, drop_first = True, dtype = int)
print(data_with_dummies.head(2))
print(data_with_dummies.shape)
data_with_dummies.columns.values
```

```python
###checking the collinearity of integer columns with the target variable y
```

```python
corr_matrix = data_with_dummies[['age','balance','day','duration','campaign','pdays','previous','y_yes']].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
###There is some level of collinearity between :Duration and y_yes, and p_days and previous.
###They are not very highly correlated so I will leave it as such for now
###Heading on to calculate the variance inflation factor for all the columns now
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = data_with_dummies[['age', 'balance', 'day', 'duration', 'campaign', 'pdays',
       'previous', 'job_blue-collar', 'job_entrepreneur', 'job_housemaid',
       'job_management', 'job_retired', 'job_self-employed',
       'job_services', 'job_student', 'job_technician', 'job_unemployed',
       'marital_married', 'marital_single', 'education_secondary',
       'education_tertiary', 'default_yes', 'housing_yes', 'loan_yes',
       'contact_telephone', 'contact_unknown', 'month_aug', 'month_dec',
       'month_feb', 'month_jan', 'month_jul', 'month_jun', 'month_mar',
       'month_may', 'month_nov', 'month_oct', 'month_sep',
       'poutcome_other', 'poutcome_success', 'poutcome_unknown']]
vif = pd.DataFrame()
vif['Features'] = variables.columns
vif['VIF'] = [variance_inflation_factor(variables.values, i) for i in range(variables.shape[1])]
vif
```

```python
###age and poutcome_unknown have high VIF, Lets look at feature importance using Decision Tree Classifier, to decide what to remove from these columns
```

```python
from sklearn.tree import DecisionTreeClassifier

x = data_with_dummies.drop(columns='y_yes')
y = data_with_dummies['y_yes']

# Training a decision tree classifier
clf = DecisionTreeClassifier(random_state=13)
clf.fit(x, y)

# Extracting feature importances
feature_importances = pd.DataFrame({
    'feature': x.columns,
    'importance': clf.feature_importances_
}).sort_values(by='importance', ascending=False)

feature_importances
```

```python
###Based on the above info, age is an important feaature.
###Removing poutcome_unknown
###Removing contact_unknown because it doesn't give any insight and might not be suitable to be used in model training
```

```python
data_p1 = data_with_dummies.drop(['contact_unknown','poutcome_unknown'], axis = 1)
data_p1.shape
```

```python
###Balancing the data by umsampling rows with y_yes = 1
```

```python
data_y_yes = data_p1[data_p1['y_yes']==1]
data_y_no = data_p1[data_p1['y_yes']==0]
print(data_y_yes.shape)
print(data_y_no.shape)
```

```python
dp_factor = len(data_y_no)//len(data_y_yes)
data_y_yes_duplicated = pd.concat([data_y_yes]*dp_factor, ignore_index = True)
balanced_data = pd.concat([data_y_no, data_y_yes_duplicated], ignore_index = True)
balanced_data_yes = balanced_data[balanced_data['y_yes'] == 1]
balanced_data_no = balanced_data[balanced_data['y_yes'] == 0]
print(len(balanced_data_yes))
print(len(balanced_data_no))
print(balanced_data.shape)
```

```python
data_final = balanced_data.copy()
```

```python
###Let's move towards building the logistic regression model now
```

```python
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
x = data_final.drop('y_yes', axis = 1)
y = data_final['y_yes']
```

```python
###Standardizing the dataset
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size = 0.2, random_state = 13)
```

```python
x_train_const = sm.add_constant(x_train)
x_test_const = sm.add_constant(x_test)
```

```python
logreg = sm.Logit(y_train, x_train_const)
results = logreg.fit()
```

```python
results.summary()
```

```python
###Not quite readable, lets make this better
```

```python
summary_df = pd.DataFrame({'Coefficients': results.params, 'P-values': results.pvalues, 'Standard Errors': results.bse})
cols1 = x.columns.values
cols2 = np.insert(cols1, 0, 'const')
cols2
summary_df['variables'] = cols2
summary_df.reset_index(drop = True)
cols = summary_df.columns.tolist()
cols = [cols[-1]]+cols[:-1]
summary_df=summary_df[cols]
summary_df['Odds'] = np.exp(summary_df['Coefficients'])
summary_df.sort_values(by = 'Odds', ascending = False)
```

```python
###Lets test this now
```

```python
y_pred_prob = results.predict(x_test_const)
y_pred = (y_pred_prob>=0.5).astype(int)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
conf_mat = confusion_matrix(y_test, y_pred)
print(conf_mat)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(conf_mat), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
acc_1 = conf_mat[1,1]/(conf_mat[1,0]+conf_mat[1,1])
acc_0 = conf_mat[0,0]/(conf_mat[0,0]+conf_mat[0,1])
print(acc_0)
print(acc_1)
```

```python
###i) This means that for each additional second of call duration, the odds of a client subscribing to a term deposit increase by a factor of 6.09
###Higher time spent on engagement with the right customer will very likely improve the chances of them subscribing to the deposit
```

```python
###If a customer gets tertiary education, the odds of subscribing to the deposit increase by 37%
```

```python
###If a customer is contacted on Mar, Oct and Sep, their odds of subscribing to the deposit increase by 36%, 23% and 17% respectively
```

```python
###A customer who has previously responded positively to a campaign (poutcome = success) is highly likely to subscribe to the deposit plan
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
###using gridsearchCV to find the best parameters for the tree, and prune it on the best fitted parameters
```

```python
param_grid = {
    'criterion': ['gini','entropy'],
    'max_depth': [10,12,15],
    'min_samples_split': [2,5,10],
    'min_samples_leaf':[1,2,5]
}
```

```python
grid_search = GridSearchCV(estimator=DecisionTreeClassifier(),
                           param_grid=param_grid,
                           cv=5, 
                           scoring='accuracy',
                           verbose=1)  
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Estimator: ", grid_search.best_estimator_)
```

```python
best_model = grid_search.best_estimator_
```

```python
y1_pred = best_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
test_accuracy = accuracy_score(y_test, y1_pred)
print(test_accuracy)
print(confusion_matrix(y_test,y1_pred))
```
